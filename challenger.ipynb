```python
import pandas as pd
```

```python
dict_invoicing = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
# criando dataframe
df_invoicing = pd.DataFrame.from_dict(dict_invoicing)
```

```python
# capturando media de valores 
average = int(df_invoicing['valor'].mean())
average
```

```python
#grafico de barras vertical
df_invoicing['data_ref'] = pd.to_datetime(df_invoicing['data_ref'])

df_invoicing['mes_ano'] = df_invoicing['data_ref'].dt.strftime('%b-%Y')

df_invoicing.sort_values('data_ref', ascending=True).plot.bar(x='mes_ano', y='valor')
```

```python
df_invoicing.sort_values('data_ref', ascending=True).plot.line(x='mes_ano', y='valor')
```
